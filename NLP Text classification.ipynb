# Bag of words

```python
for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] =  re.sub(r'\W'," ",dataset[i])
    dataset[i] =  re.sub(r'\s+'," ",dataset[i])    
```

```python
word_count = {}
```

```python
for j in dataset:
    words =  wt(j)
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
```

```python
freq_words = heapq.nlargest(100,word_count,key = word_count.get)
```

```python
final = []
for n in freq_words:
    vector = []
    for sen in dataset:
        if n in wt(sen):
            vector.append(1)
        else:
            vector.append(0)
    final.append(vector)
            
```

# TF IDF model

```python
# tf(term frequency) = no of occurance of word in the document / no of words in the document
# idf(inverse document frequency) = no of documents / no of documents containing the word
```

```python
dataset =  st(para)
for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] =  re.sub(r'\W'," ",dataset[i])
    dataset[i] =  re.sub(r'\s+'," ",dataset[i])    
for j in dataset:
    words =  wt(j)
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
freq_words = heapq.nlargest(100,word_count,key = word_count.get)            
```

```python
#idf
di = {}
for k in freq_words:
    count = 0
    for sen in dataset:
        if k in wt(sen):
            count +=1
    idf = numpy.log(len(dataset)/count+1)
    di[k] = idf
```

```python
#tf
final = {}
for n in freq_words:
    vector = []
    for sen in dataset:
        freq = 0
        for w in wt(sen):
            if w == n:
                freq  += 1
        tf =  freq/len(wt(sen))
        vector.append(tf)
    final[n] = vector    
 
```

```python
fval = []
for word  in  final.keys():
    tfval = []
    for val in final[word]:
        score = val * di[word]
        tfval.append(score)
    fval.append(tfval)    
        
        
```

```python
import pandas as pd
pd.DataFrame(numpy.asarray(fval))
```

# Latent sementic analysis

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import nltk,numpy,pandas

dataset = ["The amount of polution is increasing day by day",
           "The concert was just great",
           "I love to see Gordon Ramsay cook",
           "Google is introducing a new technology",
           "AI Robots are examples of great technology present today",
           "All of us were singing in the concert",
           "We have launch campaigns to stop pollution and global warming"]
```

```python
cleaned_data = [i.lower() for i in dataset]
```

```python
vectoricer = TfidfVectorizer()
```

```python
tdata = vectoricer.fit_transform(cleaned_data)
```

```python
print(tdata[0])
```

```python
lsa = TruncatedSVD(n_components=4,n_iter=200)
```

```python
lsa.fit(tdata)
```

```python
terms = vectoricer.get_feature_names()
concept = {}
```

```python
for i,com in enumerate(lsa.components_):
    data = zip(terms,com) 
    sortedterm  = sorted(data,key= lambda x:x[1],reverse=True)
    concept[i] = sortedterm[:10]
    
```

```python
for key in concept.keys():
    sentence_scores = []
    for sentence in dataset:
        words = nltk.word_tokenize(sentence)
        score = 0
        for word in words:
            for word_with_score in concept[key]:
                if word == word_with_score[0]:
1                    score += word_with_score[1]
        sentence_scores.append(score)
    print("\n"+str(key)+":")
    for sentence_score in sentence_scores:
        print(sentence_score)
```
